```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the Dataset
df = pd.read_csv('C:/Users/marut/Desktop/Regression/Simple_Linear_Regression/Salary_Data.csv')
X = df.iloc[:, :-1].values # splitting Input features and output features
y = df.iloc[:, 1].values # change the columns according to the dataset and problem requirement
```

```python
# splitting the dataset into the training set and test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# predicting the Test Set Results
y_pred = reg.predict(X_test)
```

```python
# Visualising the training set and the results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)') 
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```

```python
# Visualizing the test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)') 
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```

```python
# Calculatinng the RMSE(Root Mean Square Error) of the model
from sklearn import metrics
y_pred = reg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Calculating The accuracy of the model
reg.score(X_test, y_test)
```
